```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esse arquivo.

```python
df_sinasc = pd.read_csv('https://raw.githubusercontent.com/rhatiro/Curso_EBAC-Profissao_Cientista_de_Dados/main/Mo%CC%81dulo%2006%20-%20Descritiva%20I/database/sinasc_RO_2019.csv')

print(df_sinasc.info())

df_sinasc
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df_sinasc.loc[(df_sinasc['APGAR5'] >= 0) & 
              (df_sinasc['APGAR5'] <= 3), 
              'APGAR5_categorizado'] = 'asfixia grave'

df_sinasc.loc[(df_sinasc['APGAR5'] >= 4) & 
              (df_sinasc['APGAR5'] <= 5), 
              'APGAR5_categorizado'] = 'asfixia moderada'

df_sinasc.loc[(df_sinasc['APGAR5'] >= 6) & 
              (df_sinasc['APGAR5'] <= 7), 
              'APGAR5_categorizado'] = 'asfixia leve'

df_sinasc.loc[(df_sinasc['APGAR5'] >= 8) & 
              (df_sinasc['APGAR5'] <= 10), 
              'APGAR5_categorizado'] = 'normal'

df_sinasc['APGAR5_categorizado'].value_counts()
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
tempo_gestacao = df_sinasc['GESTACAO'].value_counts().reindex(
    ['Menos de 22 semanas', 
     '22 a 27 semanas', 
     '28 a 31 semanas', 
     '32 a 36 semanas', 
     '37 a 41 semanas', 
     '42 semanas e mais']).index

for gestacao in tempo_gestacao:
    sns.countplot(x='APGAR5', 
                  hue='SEXO', 
                  data=df_sinasc[df_sinasc.GESTACAO == gestacao])
    plt.legend(loc='upper left')
    plt.title(label='Tempo de gestação: {}'.format(gestacao))
    plt.show()
```

```python
sns.catplot(x='GESTACAO', 
            y='APGAR5',
            hue='SEXO', 
            data=df_sinasc, 
            order=['Menos de 22 semanas', 
                   '22 a 27 semanas', 
                   '28 a 31 semanas', 
                   '32 a 36 semanas', 
                   '37 a 41 semanas', 
                   '42 semanas e mais'], 
            kind='bar')
plt.xticks(rotation=45)
plt.xlabel(xlabel='Tempo de gestação')
plt.ylabel(ylabel="Média de 'APGAR5'")

plt.show()
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tabela_cruzada = pd.crosstab(index=df_sinasc['GESTACAO'], 
                             columns=df_sinasc['APGAR5_categorizado'])

tabela_cruzada = tabela_cruzada.reindex(
    index=['Menos de 22 semanas', 
           '22 a 27 semanas', 
           '28 a 31 semanas', 
           '32 a 36 semanas', 
           '37 a 41 semanas', 
           '42 semanas e mais'], 
    columns=['asfixia grave', 
             'asfixia moderada', 
             'asfixia leve', 
             'normal'])

tabela_cruzada
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
pct_tabela_cruzada = tabela_cruzada.div(tabela_cruzada.sum(axis=1), axis=0)

print(pct_tabela_cruzada.sum(axis=1))

pct_tabela_cruzada
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
plt.rc('figure', figsize=(12,6))

pct_tabela_cruzada.plot.bar(color=['red', 'orange', 'yellow', 'green'])
plt.xticks(rotation=45)
plt.xlabel(xlabel='Tempo de gestação')
plt.legend(title='Apgar no quinto minuto')

plt.show()
```

De fato, ao analisar o gráfico acima, nota-se que, após as 37 semanas de gestação, os riscos de asfixia e baixas pontuações no teste de Apgar no quinto minuto são praticamente nulos. Observa-se também que quanto menor o tempo de gestação, maior é a probabilidade de ocorrência de asfixia.  
É importante destacar que bebês prematuros, incluindo aqueles com menos de 22 semanas de gestação, têm a possibilidade de obter boas pontuações no teste de Apgar conforme apresentado, inclusive com a pontuação 'normal' superando a quantidade de pontuações inferiores consideradas de risco.

```python
plt.rc('figure', figsize=(12,6))
fix, axes = plt.subplots(1,2)

sns.barplot(x='APGAR5', y='PESO', data=df_sinasc, ax=axes[0])
axes[0].set_title(label='seaborn.barplot')

sns.boxplot(x='APGAR5', y='PESO', data=df_sinasc, ax=axes[1])
axes[1].set_title(label='seaborn.boxplot')

plt.show()
```

Ao relacionar as duas variáveis, observa-se que quanto maior a pontuação de Apgar, menor é a variação de peso. Em contrapartida, baixas pontuações de Apgar podem indicar um possível risco de peso abaixo do normal. Além disso, percebe-se que à medida que a variação de peso se estabiliza até a pontuação máxima, aumenta a quantidade de valores discrepantes em níveis altos e baixos, o que evidencia a necessidade de atenção individualizada.
