```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df_sinasc = pd.read_csv('https://raw.githubusercontent.com/rhatiro/Curso_EBAC-Profissao_Cientista_de_Dados/main/Mo%CC%81dulo%2006%20-%20Descritiva%20I/database/sinasc_RO_2019.csv')
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
plt.rc('figure', figsize=(12,6))

fig, axes = plt.subplots(1,2)

# Utilizando a biblioteca pandas:
df_sinasc.boxplot(column='IDADEMAE', ax=axes[0])
axes[0].set_title(label="'IDADEMAE'")
axes[0].set_ylabel(ylabel='Idade da mãe (em anos)')
axes[0].set_xticks(ticks=[1], labels='')

# Utilizando a biblioteca seaborn:
sns.boxplot(y='IDADEMAE', data=df_sinasc, ax=axes[1])
axes[1].set_title(label="'IDADEMAE'")
axes[1].set_ylabel(ylabel='Idade da mãe (em anos)')

plt.subplots_adjust(wspace=.3)
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.rc('figure', figsize=(12,6))

fig, axes = plt.subplots(1,2)

# Utilizando a biblioteca pandas:
df_sinasc.boxplot(column='PESO', ax=axes[0])
axes[0].set_title(label="'PESO'")
axes[0].set_ylabel(ylabel='Peso ao nascer (em gramas)')
axes[0].set_xticklabels([''])

# Utilizando a biblioteca seaborn:
sns.boxplot(y='PESO', data=df_sinasc, ax=axes[1])
axes[1].set_title(label="'PESO'")
axes[1].set_ylabel(ylabel='Peso ao nascer (em gramas)')

plt.subplots_adjust(wspace=.3)
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.rc('figure', figsize=(12,5))

fig, axes = plt.subplots(1,2)

# Utilizando a biblioteca pandas:
df_sinasc['IDADEMAE'].plot.hist(ax=axes[0])
axes[0].set_title(label="'IDADEMAE'")
axes[0].set_xlabel(xlabel='Idade da mãe (em anos)')

# Utilizando a biblioteca seaborn:
sns.histplot(data=df_sinasc['IDADEMAE'], bins=10, ax=axes[1])
axes[1].set_title(label="'IDADEMAE'")
axes[1].set_xlabel(xlabel='Idade da mãe (em anos)')

plt.subplots_adjust(wspace=.3)
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
plt.rc('figure', figsize=(12,5))

fig, axes = plt.subplots(1,2)

# Utilizando a biblioteca pandas:
df_sinasc['PESO'].plot.hist(bins=30, ax=axes[0])
axes[0].set_title(label="'PESO'")
axes[0].set_xlabel(xlabel='Peso ao nascer (em gramas)')

# Utilizando a biblioteca seaborn:
sns.histplot(data=df_sinasc['PESO'], bins=30, ax=axes[1])
axes[1].set_title(label="'PESO'")
axes[1].set_xlabel(xlabel='Peso ao nascer (em gramas)')

plt.subplots_adjust(wspace=.3)
plt.show()
```
